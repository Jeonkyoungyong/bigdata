```python
#json Data 처리

import pandas as pd
import json

#Json 파일 경로
json_file_path = "/content/drive/MyDrive/빅데이터 9주차/과제/pokemonDB_dataset.json"

#json 파일 읽기
with open (json_file_path, 'r', encoding='utf-8') as f:
  json_data = json.load(f)

#json 데이터를 pandas Dataframe으로 변환
df = pd.DataFrame(json_data)

# Dataframe 정보 출력
print(df.info())

#Dataframe 상위 5개 출력
print(df.head())
```

```python
#엑셀데이터

import pandas as pd
import xml.etree.ElementTree as ET

#XML파일경로
xml_file_path = "/content/drive/MyDrive/빅데이터 9주차/xmlData/book.xml"

#XML 파일 읽기
tree = ET.parse(xml_file_path)

#루트요소 가져오기
root = tree.getroot()

#특정요소 목록 가져오기
elements = root.findall('book')

#dataframe 생성
df = pd.DataFrame(columns=['title', 'author', 'price', 'genre'])

#각ㅇ소를 Dataframe에 추가
for element in elements:
  row = {
      'title' : element.find('title').text,
      'author' : element.find('author').text,
      'price' : float(element.find('price').text),
      'genre' : element.find('genre').text,
  }
  new_df = pd.DataFrame([row], columns=df.columns)
  df = pd.concat([df, new_df], ignore_index=True)

  #DataFrame 정보 출력
  print(df.info())

  #DataFrame 상위 5개 행 출력
  print(df.head())

```
