```python
#엑셀데이터

import pandas as pd
import xml.etree.ElementTree as ET

#XML파일경로
xml_file_path = "/content/drive/MyDrive/빅데이터 9주차/과제/countries+states.xml"

#XML 파일 읽기
tree = ET.parse(xml_file_path)

#루트요소 가져오기
root = tree.getroot()

#특정요소 목록 가져오기
elements = root.findall('country_state')

#dataframe 생성
df = pd.DataFrame(columns=['name', 'capital', 'currency', 'region'])

#각ㅇ소를 Dataframe에 추가
for element in elements:
  row = {
      'name' : element.find('name').text,
      'capital' : element.find('capital').text,
      'currency' : element.find('currency').text,
      'region' : element.find('region').text,
  }
  new_df = pd.DataFrame([row], columns=df.columns)
  df = pd.concat([df, new_df], ignore_index=True)

  #DataFrame 정보 출력
  print(df.info())

  #DataFrame 상위 5개 행 출력
  print(df.head())

```
