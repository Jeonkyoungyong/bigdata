```python
import pandas as pd
file = '/content/drive/MyDrive/11week/owid-covid-data-2.csv'
df = pd.read_csv(file)
df.info()

df = pd.read_csv(file)
wanted_df = pd.read_csv(file)
df.info()
```

```python
#대한민국 데이터 추출하기
KOR_df = wanted_df[wanted_df.location == 'South Korea']
KOR_df.head()
```

```python
#데이터 프레임 색인 변경하기 : set_index()

KOR_date_index_df = KOR_df.set_index('date')
KOR_date_index_df.head()
```

```python
#프랑스 데이터, 날짜 색인 만들기
FRA_df = wanted_df[wanted_df.location == 'France']
FRA_df.head()

FRA_date_index_df = FRA_df.set_index('date')
FRA_date_index_df.head()
```

```python
#대한민국 확진자 수 구하기
KOR_total_cases = KOR_date_index_df['total_cases']
KOR_total_cases
```

```python
#프랑스 확진자 수 구하기
FRA_total_cases = FRA_date_index_df['total_cases']
FRA_total_cases
```

```python
#데이터를 그리기 위한 최종 데이터 프레임 만들기
target_df = pd.DataFrame({
    'KOR' : KOR_total_cases,
    'FRA' : FRA_total_cases
  }, index = KOR_date_index_df.index
)
target_df.head()
```

```python
#인구대비 확진자 비율 구하기 : 데이터 프레임 population행의 2022년 1월 5일자 인구를 확인
KOR_population = KOR_date_index_df['population']['2020-02-05']
FRA_population = FRA_date_index_df['population']['2020-02-05']
print('프랑스인구(',FRA_population, ') / 대한민국 인구(',
      KOR_population, ')')
```

```python
#인구대비 확진자 비율 구하기 : 프랑스 인구를 대한민국 인구로 나눠 비율 확인
rate = round((FRA_population / KOR_population), 2)
rate
```

```python
#확진자 비율 반영한 데이터 프레임 생성하기 : 인구비율 X 대한민국 확진자수 / 실행결과 1.31배 비율 반영
target_wanted_df = pd.DataFrame({
    'KOR' : KOR_total_cases * rate,
    'FRA' : FRA_total_cases
}, index = KOR_date_index_df.index
)

target_wanted_df.iloc[50:56]
```

```python
#실행 결과
target_wanted_df[:].plot.line(rot=45)
```
